```python
from audiocraft.utils import export
from audiocraft import train
import audiocraft
from audiocraft.utils.notebook import display_audio

import soundfile as sf
import torch
import numpy as np
import random
```

```python
def setup_seed(seed):
     torch.manual_seed(seed)
     torch.cuda.manual_seed_all(seed)
     np.random.seed(seed)
     random.seed(seed)
     torch.backends.cudnn.deterministic = True
```

```python
def import_test(prompts, name):
    """
    Import a finetuned model and generate audio, then display the first audio
    """
    
    setup_seed(42)
    
    musicgen = audiocraft.models.MusicGen.get_pretrained(f'./checkpoints/{name}/')
    musicgen.set_generation_params(
        use_sampling=True,
        top_k=250,
        duration=10
    )

    output = musicgen.generate(
        descriptions=prompts,
        progress=True, return_tokens=True
    )
    display_audio(output[0], sample_rate=32000)
    return output[0]


def test_original(prompts):
    """
    Import an original model (small) and generate audio, then display the first audio
    """
    setup_seed(42)
    
    musicgen = audiocraft.models.MusicGen.get_pretrained('small')
    musicgen.set_generation_params(
        use_sampling=True,
        top_k=250,
        duration=10
    )

    output = musicgen.generate(
        descriptions=prompts,
        progress=True, return_tokens=True
    )
    display_audio(output[0], sample_rate=32000)
    return output[0]
    
```
